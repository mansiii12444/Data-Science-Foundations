```python
import joblib
```

```python
LE=joblib.load("LabelE.pkl")
LOGR=joblib.load("LOGreg.pkl")
```

```python
['Gender', 'Study_Hours', 'Attendance', 'Previous_Score','Final_Score']
```

```python
GENDER = input("Write your gender -- Male,Female -- ")
Gender=LE.fit_transform([GENDER])
print(Gender)
Study_Hours=int(input("Write your study hours"))
Attendance=int(input("Write your attendance"))
Previous_Score=int(input("Write your previous score"))
Final_Score=int(input("Write your final score"))
print('Gender', 'Study_Hours', 'Attendance', 'Previous_Score','Final_Score')
```
